###  Logical Operators
Logical operators are used to combine conditional statements.

```python
x = 9
y = 10
```

```python
# not
print(not(x > 6)) # Reverses the result (negates the condition).
```

### Bitwise Operators
Bitwise operators work on binary (bit) level - they operate directly on the 0s and 1s of numbers.

```python
x = 10
y = 3
```

```python
bin(10) # binary figure for 10
```

```python
bin(3) # binary figure for 3
```

```python
# Bitwise AND (&) - Both must be 1 to get 1.
#   1010 (10)
# & 0011 (3)
#   0010
# We then convert 10 (base 2) to decimal (base 10)
# (1*2**1)+(0*2**0)
# (1*2)+(0*1)
# 2+0
# 2
x & y
```

```python
# Bitwise OR (|) - If at least one is 1 → result is 1.
#   1010
# | 0011
#   1011
# We then convert 1011 (base 2) to decimal (base 10)
# (1*2**3)+(0*2**2)+(1*2**1)+(1*2**0)
# (1*8)+(0*4)+(1*2)+(1*1)
# 8+0+2+1
# 11
x|y
```

```python
# Bitwise XOR (^) - It returns 1 only if the bits are different.
#   1010
# ^ 0011
#   1001
# We then convert 10 (base 2) to decimal (base 10)
# (1*2**3)+(0*2**2)+(0*2**1)+(1*2**0)
# (1*8)+(0*4)+(0*2)+(1*1)
# 8+0+0+1
# 9
x^y
```

```python
# Bitwise NOT (~) - This flips all bits.
#  ~x = -(x + 1)
# ~10 = -11
~x
```

```python
# right Shift (>>) - Each shift divides by 2 (floor division for integers)
# Effect: Move all bits n positions to the right.
# Think of right shift as “dropping the last n bits”.
# 10/2
# 5/2
# 2
print(x >> 2)
```

```python
# left Shift (<<) -  Each shift multiplies by 2.
# Effect: Move all bits n positions to the left.
# Think of left shift as “adding zeros to the right” like multiplying by 10 in decimal, but in base 2.
# 10*2
# 20*2
# 40*2
# 80
print(x << 3)
```

## Identity Operators
Python has two identity operators:
1. is
2. is not

They check if two objects are the same object in memory, not just equal in value.

```python
a = [1,2,3,4,5]
b = [1,2,3,4,5]                           
```

```python
# print(a==b) # == → checks if values are equal 
print( a is b) #is → checks if they are the same object in memory                         
```

```python
print( a is not b)
```
