#### Data Types
Python has several built-in data types:
1. Numeric: int, float, complex
2. Sequence: list, tuple, range
3. Mapping: dt: str

#### Numeric
##### Store numbers: int = whole numbers, float = decimals, complex = real + imaginary

```python
x = 2
y = 3.14
z = 2 + 3j
```

```python
print(type(x))
```

```python
print(type(y))
```

```python
print(type(z))
```

#### Sequence
##### Ordered collections of items; list = mutable, tuple = immutable, range = sequence of numbers

```python
# Use underscore (_) to avoid conflicts with Python Keywords or Built-in functions
list_ = [1,2,3,4,5] 
tuple_ = (1,2,3,4,5)
range_ = range(1,5)
```

```python
print(type(list_))
```

```python
print(type(tuple_))
```

```python
print(type(range_))
```
