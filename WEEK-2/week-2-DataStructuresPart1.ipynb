### LISTS
#### Lists are ordered collections of items in Python. They are mutable, meaning their elements can be modified after creation.  They  are created using square brackets [] or the list() constructor.

#### Common List Methods

```python
# Creating a list
```

```python
my_list = [1,2,3,4,5] #Lists can also be created using the list() constructor: my_list = list([1, 2, 3, 4, 5])
```

##### append(): Adds an element to the end of the list.

```python
my_list.append(6)
print(my_list)
```

##### extend(): Extends the list by appending elements from another list.

```python
my_list.extend([7,8,9,10])
print(my_list)
```

##### insert(): Inserts an element at a specified position.

```python
my_list.insert(1,1) # start with position and what you need inserted
print(my_list)
```

##### remove(): Removes the first occurrence of a specified value.

```python
my_list.remove(1) # You can only remove one 1 item at a time from the list
print(my_list)
```

##### pop(): Removes and returns the element at a specified position (default is the last element).

```python
popped_element=my_list.pop(5)
print(popped_element)
print(my_list)
```

##### index(): Returns the index of the first occurrence of a specified value.

```python
index = my_list.index(10)
index
```

##### count(): Returns the number of occurrences of a specified value.

```python
count = my_list.count(9)
print(count)
```

##### sort(): Sorts the list in ascending order.

```python
my_list.sort()
my_list
```

##### reverse(): Reverses the order of elements in the list.

```python
my_list.reverse() # sort first before reverse
my_list
```

### TUPLE
##### A tuple is just like a list BUT immutable (cannot change after creation).
Tuples are created using parentheses () or the tuple() constructor.

#### Common Tuple Methods

```python
# Creating a tuple
my_tuple = ('milk','sugar','cinnamon','oats','sugar')
```

##### count(): Returns how many times a value appears

```python
count = my_tuple.count('sugar')
print(count)
```

##### index():Returns position of the first occurrence

```python
index = my_tuple.index('oats')
print(index)
```

```python
 print(my_tuple[2]) #prints 3rd item in a tuple
```

### DICTIONARY
#### Dictionaries are unordered collections of key-value pairs. They are mutable and are useful for storing data in the form of key-value mappings.
Dictionaries are created using curly braces {} or the dict() constructor.

```python
# Creating a dictionary
my_dict = {'name':'Vicky','age':29,'city':'Nairobi'} # or using constructor: my_dict = dict(name='Vicky', age=29, city='Nairobi')
```

##### Common Dictionary Methods

##### get(): Returns the value for a specified key.

```python
age = my_dict.get('age')
print(age)
```

##### indexing : Indexing in dictionaries means accessing a value using its key.

```python
# indexing[] In a dictionary, the square brackets [] are used to access a value using its key (NOT position).
age = my_dict['age']
print(age)
```

##### items(): Returns a view object containing key-value pairs.

```python
items = my_dict.items()
print(items)
```

##### keys(): Returns a view object containing the keys.

```python
keys = my_dict.keys()
keys
```

##### values(): Returns a view object containing the values.

```python
values = my_dict.values()
values
```

##### update(): Updates the dictionary with the specified key-value pairs from another dictionary or iterable.

```python
my_dict.update({'age':30,'gender':'female','country':'Kenya'})
my_dict
```
